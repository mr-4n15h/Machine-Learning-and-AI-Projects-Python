```python
# Feature engineering process
# Right now the data contains a lot of features which can be engineered

data.head()
```

```python
# Let's remove the id column because it is not really helpful

data = data.drop('id', axis=1)
data.head()
```

```python
# Let's feature engineer the date feature
# Right now, the date column is of type object

data['date'] = pd.to_datetime(data['date'])
data.head()
```

```python
# Extract the month and year from the date
data['year'] = data['date'].apply(lambda date: date.year)
data['month'] = data['date'].apply(lambda date: date.month)
data.head()
```

```python
# Check the price overall distribution
# This plot is interesting because the price of the houses tend to increase after the second month
data.groupby('month').mean()['price'].plot()
```

```python
# This plot is obvious because houses tend to get more expensive as time goes in
data.groupby('year').mean()['price'].plot()
```

```python
# Hmmm, the correlation below show there is very little correlation between the features and price
# But, let's still use them
data[['year', 'month', 'price']].corr()
```

```python
# Drop the date column because we already feature engineered of it
data.drop('date', inplace=True, axis=1)
```

```python
data.head()
```

```python
# Zipcode feature does not really help in predicting houses
# Therefore, we can remove this feature! We will use the lat and long features to predict houses based on location instead
data['zipcode'].value_counts()
```

```python
data.drop('zipcode', inplace=True, axis=1)
```

```python
data.head()
```

```python
X = data.drop('price', axis=1).values
y = data['price'].values
```

```python
# Split the data and scale them

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=19, activation="relu"))
model.add(Dense(units=19, activation="relu"))
model.add(Dense(units=19, activation="relu"))
model.add(Dense(units=19, activation="relu"))
model.add(Dense(units=1))

model.compile(optimizer="adam", loss="mse")

model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

```python
model_history = model.history.history
model_history
```

```python
model_history_df = pd.DataFrame(data=model_history)
model_history_df
```

```python
model_history_df.plot(figsize=(15, 10))
```

```python
model_history_df.plot(x="loss", y="val_loss", kind="line")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import explained_variance_score

y_pred = model.predict(X_test)

print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred) ** .5)
```

```python
# the model is OK - not too good, not too bad, just somewhere between
explained_variance_score(y_test, y_pred)
```
