```python
# Visualisation is difficult because the data has been transformed using PCA
# therefore, we do not know what the features are
# Let's begin the training

# Scale the Amount feature 

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data['Amount_normalised'] = scaler.fit_transform(data['Amount'].values.reshape(-1, 1))
X = data.drop(['Class', 'Amount'], axis=1)
y = data['Class']
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# OK model, but it has a hard time predicting data 

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
