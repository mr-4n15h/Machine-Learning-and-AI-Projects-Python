```python
data.drop(['date', 'rating'], axis=1, inplace=True)
data.head()
```

```python
variation_dummies = pd.get_dummies(data['variation'], drop_first=True)
variation_dummies
```

```python
data.drop(['variation'], axis=1, inplace=True)
data.head()
```

```python
data = pd.concat([variation_dummies, data], axis=1)
data.head()
```

```python
# Count vectorise the reviews
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
reviews_countvectorizer = vectorizer.fit_transform(data['verified_reviews'])
```

```python
reviews_countvectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
data.head()
```

```python
data.drop(["verified_reviews"], axis=1, inplace=True)
```

```python
data.head()
```

```python
encoded_reviews = pd.DataFrame(reviews_countvectorizer.toarray())
data = pd.concat([data, encoded_reviews], axis=1)
```

```python
data.head()
```

```python
data.shape
```

```python
# Training and testing data preparation
X = data.drop(['feedback'], axis=1)
X
```

```python
y = data['feedback']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=100, criterion="entropy")
classifier.fit(X_train, y_train)
```

```python
y_preds = classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_preds))
print()
print(accuracy_score(y_test, y_preds))
print()
print(confusion_matrix(y_test, y_preds))
```
