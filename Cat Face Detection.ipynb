```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
catface_cascade_ext = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml')
catface_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalcatface.xml")
```

```python
def detect(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces_ext = catface_cascade_ext.detectMultiScale(gray, 1.3, 10, minSize=(10,10))
    faces = catface_cascade.detectMultiScale(gray, 1.15, 1, minSize=(10,10))
    for (i, (x, y, w, h)) in enumerate(faces_ext):
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
        cv2.putText(image, "Cat #{}".format(i + 1), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (0, 0, 255), 2)
    for (i, (x, y, w, h)) in enumerate(faces):
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 255, 0), 2)
        cv2.putText(image, "Cat #{}".format(i + 1), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (0, 0, 255), 2)
    return image
```

```python
cat1_color = plt.imread('Faces/cats/cat.17.jpg')
cat1_test_detect = detect(cat1_color)
plt.imshow(cat1_test_detect)
```
