```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 5)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 0, 255), 2)
    return frame
```

```python
image = plt.imread('Faces/genova.jpg')
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray")
```

```python
test_example = detect(gray_image, image)
plt.imshow(test_example)
```

```python
female_color_image = plt.imread('Faces/female.jpg')
female_color_image = cv2.resize(female_color_image, (500, 500))
female_gray_image = cv2.cvtColor(female_color_image, cv2.COLOR_BGR2GRAY)
plt.imshow(female_gray_image, cmap="gray")
```

```python
test_example = detect(female_gray_image, female_color_image)
plt.imshow(test_example)
```

```python
faces_color_image = plt.imread('Faces/fake_ai_faces.jpg')
faces_gray_image = cv2.cvtColor(faces_color_image, cv2.COLOR_BGR2GRAY)
plt.imshow(faces_gray_image, cmap="gray")
```

```python
test_faces = detect(faces_gray_image, faces_color_image)
plt.imshow(test_faces)
```
