```python
# We feature we are trying to predict is in a categorical string format
# Therefore, we need to fix the data first
# Let's drop the college names first, because they do not really help with prediction
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head()
```

```python
# Perfect, we changed the categorical variable by getting its dummy values
data = pd.get_dummies(data, drop_first=True)
data.head()
```

```python
X = data.drop('Private_Yes', axis=1).values

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)
model.fit(X)
model.cluster_centers_
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(data['Private_Yes'], model.labels_))
print(classification_report(data['Private_Yes'], model.labels_))
```

```python
# The model is quite poor
# We can use this elbow technique, but we know there are only two clusters 
# in this dataset, so there is no point

distances = []
K = range(1, 30)
for k in K:
    model = KMeans(n_clusters=k)
    model = model.fit(X)
    distances.append(model.inertia_)

plt.figure(figsize=(10, 10))
plt.plot(K, distances, 'bx-')
plt.xlabel("K")
plt.ylabel("Sum of squared distances")
plt.show()
```
